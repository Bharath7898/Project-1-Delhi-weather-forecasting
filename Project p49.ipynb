# Forecasting_Model based methods


```python
data.month.value_counts()
```

```python
data['month']=data['date'].dt.month
data.head()
```

```python
month_dummies = pd.DataFrame(pd.get_dummies(data['month']))
data = pd.concat([data,month_dummies],axis = 1)
data.head()
```

```python
data.rename(columns = {1:'jan',2:'feb',3:'mar',4:'apr'}, inplace = True)
```

```python
data["t"]=np.arange(1,2375)
data.head()
```

```python
data["t_squared"]=data.t*data.t
data["log_pm25"]=np.log(data.pm25)
data.head()
```

```python
#splitting data into train and test data.
Train = data.head(2050)
Test = data.tail(324)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.formula.api as smf
from statsmodels.tsa.holtwinters import SimpleExpSmoothing #SES
from statsmodels.tsa.holtwinters import Holt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
# Linear model
linear_model = smf.ols('pm25~t',data=Train).fit()
pred_linear =  pd.Series(linear_model.predict(pd.DataFrame(Test['t'])))
rmse_linear = np.sqrt(np.mean((np.array(Test['pm25'])-np.array(pred_linear))**2))
rmse_linear
```

```python
# Exponential
Exp = smf.ols('log_pm25~t',data=Train).fit()
pred_Exp = pd.Series(Exp.predict(pd.DataFrame(Test['t'])))
rmse_Exp = np.sqrt(np.mean((np.array(Test['pm25'])-np.array(np.exp(pred_Exp)))**2))
rmse_Exp
```

```python
# Quadratic
Quad = smf.ols('pm25~t+t_squared',data=Train).fit()
pred_Quad = pd.Series(Quad.predict(Test[["t","t_squared"]]))
rmse_Quad = np.sqrt(np.mean((np.array(Test['pm25'])-np.array(pred_Quad))**2))
rmse_Quad
```

```python
# Additive seasonality
add_sea = smf.ols('pm25~jan+feb+mar+apr',data=Train).fit()
pred_add_sea = pd.Series(add_sea.predict(Test[['jan','feb','mar','apr']]))
rmse_add_sea = np.sqrt(np.mean((np.array(Test['pm25'])-np.array(pred_add_sea))**2))
rmse_add_sea
```

```python
# Additive Seasonality Quadratic
add_sea_Quad = smf.ols('pm25~t+t_squared+jan+feb+mar+apr',data=Train).fit()
pred_add_sea_quad = pd.Series(add_sea_Quad.predict(Test[['jan','feb','mar','apr','t','t_squared']]))
rmse_add_sea_quad = np.sqrt(np.mean((np.array(Test['pm25'])-np.array(pred_add_sea_quad))**2))
rmse_add_sea_quad 
```

```python
# Multiplicative Seasonality
Mul_sea = smf.ols('log_pm25~jan+feb+mar+apr',data = Train).fit()
pred_Mult_sea = pd.Series(Mul_sea.predict(Test[['jan','feb','mar','apr']]))
rmse_Mult_sea = np.sqrt(np.mean((np.array(Test['pm25'])-np.array(np.exp(pred_Mult_sea)))**2))
rmse_Mult_sea
```

```python
# Multiplicative Additive Seasonality
Mul_Add_sea = smf.ols('log_pm25~t+jan+feb+mar+apr',data = Train).fit()
pred_Mult_add_sea = pd.Series(Mul_Add_sea.predict(Test[['jan','feb','mar','apr','t']]))
rmse_Mult_add_sea = np.sqrt(np.mean((np.array(Test['pm25'])-np.array(np.exp(pred_Mult_add_sea)))**2))
rmse_Mult_add_sea
```

```python
#Compare the results 

datac = {"MODEL":pd.Series(["rmse_linear","rmse_Exp","rmse_Quad","rmse_add_sea","rmse_add_sea_quad","rmse_Mult_sea","rmse_Mult_add_sea"]),"RMSE_Values":pd.Series([rmse_linear,rmse_Exp,rmse_Quad,rmse_add_sea,rmse_add_sea_quad,rmse_Mult_sea,rmse_Mult_add_sea])}
table_rmse=pd.DataFrame(datac)
table_rmse.sort_values(['RMSE_Values'])
```

# Forecasting_Data Driven Models

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.holtwinters import SimpleExpSmoothing # SES
from statsmodels.tsa.holtwinters import Holt # Holts Exponential Smoothing
from statsmodels.tsa.holtwinters import ExponentialSmoothing 
```

```python
df = pd.read_excel(r"C:\Users\DELL\Desktop\bharath's file\DS\csv\delhi1.xlsx")
df
```

```python
df.pm25 = pd.to_numeric(df.pm25, errors = 'coerce')
df.pm25.plot()
```

```python
#splitting data into train and test data.
Train = df.head(2050)
Test = df.tail(324)
```

```python
#Moving Average
plt.figure(figsize=(12,4))
df.pm25.plot(label="org")
for i in range(2,24,6):
    df["pm25"].rolling(i).mean().plot(label=str(i))
plt.legend(loc='best')
```

```python
df.mean()
df.pm25.fillna(df['pm25'].mean(),inplace=True)
```

```python
# Time series decomposition plot

decompose_ts_add = seasonal_decompose(df.pm25,period=12)
decompose_ts_add.plot()
plt.show()
```

```python
# ACF plots and PACF plots
import statsmodels.graphics.tsaplots as tsa_plots
tsa_plots.plot_acf(df.pm25,lags=12)
tsa_plots.plot_pacf(df.pm25,lags=12)
plt.show()
```

```python
# Evaluation Metric MAPE
def MAPE(pred,org):
    temp = np.abs((pred-org)/org)*100
    return np.mean(temp)
```

```python
# Simple Exponential Method
ses_model = SimpleExpSmoothing(Train['pm25']).fit(smoothing_level=0.2)
pred_ses = ses_model.predict(start = Test.index[0],end = Test.index[-1])
MAPE(pred_ses,Test.pm25) 
```

```python
# Holt method 
hw_model = Holt(Train["pm25"]).fit(smoothing_level=0.8, smoothing_slope=0.2)
pred_hw = hw_model.predict(start = Test.index[0],end = Test.index[-1])
MAPE(pred_hw,Test.pm25) 
```

```python
# Holts winter exponential smoothing with additive seasonality and additive trend
hwe_model_add_add = ExponentialSmoothing(Train["pm25"],seasonal="add",trend="add",seasonal_periods=12).fit() #add the trend to the model
pred_hwe_add_add = hwe_model_add_add.predict(start = Test.index[0],end = Test.index[-1])
MAPE(pred_hwe_add_add,Test.pm25) 
```

```python
# Holts winter exponential smoothing with multiplicative seasonality and additive trend
hwe_model_mul_add = ExponentialSmoothing(Train["pm25"],seasonal="mul",trend="add",seasonal_periods=12).fit() 
pred_hwe_mul_add = hwe_model_mul_add.predict(start = Test.index[0],end = Test.index[-1])
MAPE(pred_hwe_mul_add,Test.pm25)
```

```python
#Compare the results 

datad = {"MODEL":pd.Series(["MAPE(pred_ses,Test.pm25)","MAPE(pred_hw,Test.pm25) ","MAPE(pred_hwe_add_add,Test.pm25) ","MAPE(pred_hwe_mul_add,Test.pm25)"]),"MAPE_Values":pd.Series([MAPE(pred_ses,Test.pm25),MAPE(pred_hw,Test.pm25) ,MAPE(pred_hwe_add_add,Test.pm25) ,MAPE(pred_hwe_mul_add,Test.pm25)])}
table_rmse=pd.DataFrame(datad)
table_rmse.sort_values(['MAPE_Values'])
```

```python
# Final Model by combining train and test
ses_model = SimpleExpSmoothing(df['pm25']).fit(smoothing_level=0.2)
```

```python
#Forecasting for next 10 time periods
ses_model.forecast(10)
```
